```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("/Users/adityadilipkumarwagh/Desktop/DMBIA.csv")

colors = ['red','red','red','red','red','red','red','green','green','green', 'green', 'green', 'green','green', 'blue', 'blue', 'blue', 'blue', 'blue', 'blue', 'blue', 'blue']

fig, ax = plt.subplots(figsize=(15,10))

bp = ax.boxplot(df, patch_artist=True)

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

ax.set_title("Boxplots")
ax.set_xlabel("Year")
ax.set_ylabel("GDP")

ax.set_xticklabels([i for i in df.columns])

plt.show()
```

```python
colors = ['red','red','red','red','red','red','red','green','green','green', 'green', 'green', 'green','green', 'blue', 'blue', 'blue', 'blue', 'blue', 'blue', 'blue', 'blue']

# Create the figure and axes objects
fig, ax = plt.subplots(figsize=(12,15))

# Plot the boxplots on the same axes
bp = ax.boxplot(df, patch_artist=True, vert= False)
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)


# Add a title and labels for the x and y axes
ax.set_title("Boxplots")
ax.set_ylabel("Years")
ax.set_xlabel("GDP")

# Add labels for the three data sets
ax.set_yticklabels([i for i in df.columns.values])

# Show the plot
plt.show()
```
